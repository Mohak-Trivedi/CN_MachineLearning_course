```python
#creating an array of tuples, where each tuple's 1st element is words in the file, and 2nd element is the category of that 
#file. one tuple for each file, thus 2000 tuples.
documents = []
for category in movie_reviews.categories():
    for fileid in movie_reviews.fileids(category):
        documents.append((movie_reviews.words(fileid), category))
        
#viewing the first 5 elements in the documents array
documents[0:5]
```

```python
#let's shuffle the documents i.e. elements of the documents array, so that the files both categories will be mixed and
#it will become easier to do the train-test split
import random 
random.shuffle(documents)
#viewing the first 5 elements in the documents array
documents[0:5]
```

```python
#now we need to clean these files i.e. remove the stop words and do lemmatization
```

```python
nltk.download('wordnet')
nltk.download('averaged_perceptron_tagger')
nltk.download('stopwords')
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
#let's define our get_simple_pos() function to map pos from pos_tags() to pos of format required lemmatization (exactly same like we did in Basics of nltk)
from nltk.corpus import wordnet
def get_simple_pos(tag):
    if tag.startswith('J'):
        return wordnet.ADJ
    elif tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('N'):
        return wordnet.NOUN
    elif tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN
```

```python
from nltk import pos_tag
```

```python
#creating a list of stopwords and adding punctuations in that list too. We will need this list while data cleaning.
from nltk.corpus import stopwords
import string
stops = set(stopwords.words("english"))
punctuations = list(string.punctuation)
stops.update(punctuations)
stops
```

```python
#this is our main data cleaning function which will remove stop words and do lemmatization
def clean_review(words):
    output_words = []
    for w in words:
        if w.lower() not in stops: #converting into lower-case is done so that pos of all words can be identified correctly
            pos = pos_tag([w])#observe that we have passed array [w] and not a variable w, else we get pos tag for each character in a word, whereas we want pos tag of each word in a sentence.
            clean_word = lemmatizer.lemmatize(w, pos = get_simple_pos(pos[0][1]))#[0] refers to the required word's tuple and [1] gives the 1st element in that tuple which is the pos
            output_words.append(clean_word.lower())
    return output_words
```

```python
#Now go through each document in the documents array one-by-one and then clean the documents
documents = [(clean_review(document), category) for document, category in documents]
```

```python
#let's see the cleaned document
documents[0]
```

```python
#now let's create a feature set for each document in the documents array. 
#feature-set is nothing but the collection of top words in a document i.e. words having highest frequency.
#to find that:-
#step1: create an array containing all the words in the document
#step2: make frequency distribution so that we get to know frequency of each word and then we can select those with high frequency.
#step3: the chosen ones are in our feature set for that particular document
#now we need to represent the feature set in a particular format which is acceptable for the nltk classifier, the format is:
#({feature1: true or false, feature2: true or false,....,featureN: true or false},category)
#true if that feature comes into that document, else false
#this is one tuple for one document. in this way we make tuple for each document and put these tuples in a list
```

```python
#make sure that creating these feature sets should be done using only the training data and not the testing data

#as we have already shuffled the documents in the documents array, doing the train test split is pretty simple. We just choose the top 1500 as training and remaining for testing
training_documents = documents[0:1500]
testing_documents = documents[1500:]
```

```python
all_words = []
for doc in training_documents: #remember to use only training data
    all_words += doc[0] #remember doc is tuple, in which 0th element consists list of words in that document
#now we have all words in a list
```

```python
#now we have to create the frequency distribution
freq = nltk.FreqDist(all_words)
#now get the top 3000 words having highest frequency
common = freq.most_common(3000)
#notice in output that the value at 1st index of each tuple is the frequency of that word
common
```

```python
#we just need the words as features and not their frequency
features = [i[0] for i in common]
features
```

```python
#thus, we have got our features
#now is the time to represent them in the format which is suitable for nltk classifier as we had discussed above
```

```python
#if we view the documents array and see its element you can notice its format
documents[0]
#you can see that it contains all the words and then the category
```

```python
#let's create a function which gives us the feature dictionary as required in the format for nltk classifier
def get_feature_dict(words):
    current_features = {}
    words_set = set(words)#for efficiency
    #for each word in the features list we need to check whether that word comes into the document that we're dealing with
    #and if yes, then true is value for that feature in our dictionary that we're creating, else false
    for w in features:
        current_features[w] = w in words_set
    return current_features
```

```python
#let's check whether our get_feature_dict() works the way we want it work or not
output = get_feature_dict(training_documents[0][0]) #from 0th document passing the list of words, thus the 0 in 2nd [] too, 1 contains category
output
```

```python
#now features dictionary is ready, now to get the complete format ready we just need to get the category for each document
#and put create list of tuples where each tuple has 0th element our feature dictionary and 1st element the category of document
training_data = [(get_feature_dict(doc), category) for doc, category in training_documents]
```

```python
#similary we can get our testing data too
testing_data = [(get_feature_dict(doc), category) for doc, category in testing_documents]
```

```python
#now we will use nltk's inbuilt Naive-Bayes classifier for classification
from nltk import NaiveBayesClassifier
```

```python
classifier = NaiveBayesClassifier.train(training_data)
#training the classifier
```

```python
#find the accuracy of classifier on the testing data
nltk.classify.accuracy(classifier, testing_data)
```

```python
#we have recieved a decent accuracy without even tuning anything.
```

```python
from sklearn.feature_extraction.text import CountVectorizer #this will help us in acheiving our goal
```

```python
train_set = {"the sky is blue", "the sun is bright"}

#to get the features we will go through each word and get a list of most frequently occuring words and those would be our features.
#we don't have to do that manually, we'll do it this way:
#however, this function does not clean  the  data for us, we need to feed it already cleaned data, here we haven't
count_vec = CountVectorizer(max_features = 3) # go through the train_set and find the top 3 words(3 most frequent) and make them features. It converts each document(separated by commas in curly braces) in train_set into a row which represents the frequency in that document of the words selected as features.
#Thus, a lot of values in each record would be 0 as not all the words are present in all the documents. Thus, we will get a sparse matrix i.e. a matrix in which many values are 0.
#So, it's internally storing it in a most efficient manner so that it does not  need to store all those 0s
a = count_vec.fit_transform(train_set)#fit and transform. the transformed version is then stored in a
a # a is sparse matrix
```

```python
#to look at the sparse matrix stored in a:
a.todense()
```

```python
#this doesn't tell us much, so we have a function which will make things pretty
count_vec.get_feature_names() #this shows us the features which are selected
```

```python
#now let's use the CountVectorizer to convert the movie_review documents into a dataset of the format which is suitable for
#sklearn classifiers i.e. the features as given as X which is a 2d array(each 1d array within it will represent a record which will consist of values which represent the frequency of each feature in the document which is represented by that row.) and labels are given as Y which is 1d array.
#remember that we have already cleaned the data using our own defined function
```

```python
#first, let's get the document as an array of texts (i.e.array of strings) instead of as array of words.
#just like we had train_set
#and then get the categories as 1d array which we can use as Y
#this makes it easy for us to do the train-test split
```

```python
#right now we have the 'documents' which is an array of, where each tuple represents a document, it contains all the words of
#the cleaned version of that document stored in an array which is the 0th element of the tuple and the second element is the category to which that document belongs
documents[0]
```

```python
#and we want all the words in the array to combine and come out as a sentence, and we want to get all the categories in a separate array this will be Y

categories = [category for document, category in documents] #getting the Y i.e. array of categories

text_documents = [" ".join(document) for document, category in documents] #getting the X i.e  array of sentences, where each element of the array represents the sentence present in that particular document 
```

```python
#consider this e.g. to understand how  " ".join(document) works
a = ["ad", "is"]
" ".join(a) #joins the elements of array (strings) into one string and separates them by a " " (i.e. space)
```

```python
#thus, we have achieved the required X and Y, you can see how text_documents is documents converted into required format
text_documents[0]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(text_documents, categories)
```

```python
#now that we have our x_train which is just an array whose each element is a text of its corresponding document, which is 
#exactly what we want for CountVectorizer
count_vec = CountVectorizer(max_features = 2000, ngram_range = (1, 2))#see what ngram_range means in notes 
#now that we have our count vectorizer, it's time to do fit_transform on the training data, it's very important that you 
#don't use test data here, only training data
x_train_features = count_vec.fit_transform(x_train)
x_train_features
```

```python
x_train_features.todense() #this is actually the required X matrix of training data
```

```python
count_vec.get_feature_names() #words selected as features
```

```python
#Now to get x_test features we need to call transfom() on testing data not fit_transform()
x_test_features = count_vec.transform(x_test)
```

```python
x_test_features # even this is a sparse matrix because we have chosen 2000 features and we have only 500 docs in the test data, so there are many words not appeaering in many docs
```
